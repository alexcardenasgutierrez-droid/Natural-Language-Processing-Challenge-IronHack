```python
def clean_text(text):
    
    # 1. Remove special characters
    text = re.sub(r'\W', ' ', text)
    # 2. Remove numbers
    text = re.sub(r'\d', ' ', text)
    # 3. Remove all single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # 4. Remove single characters from the start
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)
    # 5. Substitute multiple spaces with single space
    text = re.sub(r'\s+', ' ', text)
    # 6. Remove prefixed 'b'
    text = re.sub(r'^b\s+', '', text)
    # 7. Convert to lowercase
    text = text.lower()
    return text.strip()
# Aplicar ao DataFrame
X_train['text_clean'] = X_train['text_clean'].apply(clean_text)
X_test['text_clean'] = X_test['text_clean'].apply(clean_text)
print("\nAfter Cleaning:")
display(X_train)
display(X_test)
```
